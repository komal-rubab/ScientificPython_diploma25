#### **Exercise 1:**
Create a function returning the first N terms of the tailor series of the function without using the sympy series function.

```python
import sympy as sp
```

```python
def taylor_series(f, x, n):
 x = sp.symbols('x')
 h = sp.symbols('h')
 exp = 0
 for i in range (1,n):
  exp = exp + (sp.diff(f, x, i))*((x-h)**i)/(sp.factorial(i))

 return exp
 

x = sp.symbols('x')
h = sp.symbols('h')
f = sp.sin(x+h)
n = int(input('Enter the number of terms you want to print'))
taylor_series(f, x, n)

  
```

#### **Exercise 4:**
Plot the recaman sequence in the following way:

Add a semi-circular arc from $a_{n}$ to $a_{n-1}$.

Alternate the arcs to be above and below the x-axis according to whether
    n is even or odd.

```python
def func_recaman(n):
 recaman = [0]
 for i in range (1,n):
  y = recaman[-1]
  x = y - i
  if x>0 and x not in recaman:
   recaman.append(x)
  else:
   x = y+i
   recaman.append(x)
   i=i+1
 return recaman

n = int(input('Enter the value of n:'))
seq = list()
seq = func_recaman(n)   
print(seq)
for i in range(n-1):
 x1, x2 = seq[i], seq[i+1]
 c = (x1+x2)/2
 r = abs(x2-x1)/2
 if i%2==0:
  theta = np.linspace(0, np.pi, 80)
 else:
  theta = np.linspace(np.pi, 2*np.pi, 80)
 x = c + r*np.cos(theta)
 y = r*np.sin(theta)
 plt.title('Recaman sequence',fontsize = 'large',fontstyle='italic',fontweight='bold')
 plt.plot(x, y)
 plt.axis('off')  
```

#### **Exercise 6:**

Plot a nice-looking 3-D torus. Do this 3 times with different styles. 

```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.cm as cm 
from mpl_toolkits.mplot3d import Axes3D

def torus_func(R, r):
 theta = np.linspace(0, 2*np.pi, 25)
 phi = np.linspace(0, 2*np.pi, 25)
 theta, phi = np.meshgrid(theta, phi)
 x = (R + r * np.cos(theta)) * np.cos(phi)
 y = (R + r * np.cos(theta)) * np.sin(phi)
 z = r * np.sin(theta)
 return x, y, z

R1 = 2
r1 = 1
x1, y1, z1 = torus_func(R1,r1)
fig = plt.figure()
fig = plt.figure(figsize=(16, 16)) 
a = fig.add_subplot(131, projection='3d')
a.plot_surface(x1, y1, z1, rstride=2,cstride=1,cmap='viridis')
a.view_init(elev=36, azim=26)
plt.show()

R2 = 7
r2 = 13
x2, y2, z2 = torus_func(R2,r2)
fig2 = plt.figure()
fig2 = plt.figure(figsize=(16, 16)) 
b = fig2.add_subplot(132, projection='3d')
b.plot_surface(x2, y2, z2, cmap="coolwarm", edgecolor="k")
plt.show()

R3 = 20
r3 = 5
x3, y3, z3 = torus_func(R3,r3)
fig3 = plt.figure()
fig3 = plt.figure(figsize=(16, 16)) 
c = fig3.add_subplot(133, projection='3d')
c.plot_surface(x3, y3, z3, cmap="plasma", edgecolor="k")
plt.show()
```
